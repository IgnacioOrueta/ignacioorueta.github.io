```python
# Write out my list of series:
          
URI = ['employmentandlabourmarket/peopleinwork/labourproductivity/timeseries/lzvb/prdy',
       "employmentandlabourmarket/peopleinwork/labourproductivity/timeseries/a4ym/prdy",
       'employmentandlabourmarket/peopleinwork/labourproductivity/timeseries/lzvd/prdy',
       'employmentandlabourmarket/peopleinwork/earningsandworkinghours/timeseries/kab9/emp',
       'employmentandlabourmarket/peopleinwork/earningsandworkinghours/timeseries/kac3/lms',
       'employmentandlabourmarket/peopleinwork/earningsandworkinghours/timeseries/kai7/emp',
       'employmentandlabourmarket/peopleinwork/earningsandworkinghours/timeseries/kai9/lms',
       'employmentandlabourmarket/peopleinwork/earningsandworkinghours/timeseries/ybuy/lms',
       'employmentandlabourmarket/peopleinwork/employmentandemployeetypes/timeseries/lf24/lms']

# Set the base url:
url_base = "https://api.beta.ons.gov.uk/v1/data?uri={}"

# Set a base fileName:
for i in ons_series:
  file_base = "../data/data_ONS-{}.json"

for i in URI:
  # Print the series code we're about to download.
  print(i)

  # Build the URL for this iteration of the loop, and check what we are getting:
  URL = url_base.format(i)
  print(URL)
```

### Batch downloader

```python
# Begin a loop, dealing with each series:
j = 0 # This will allow me to assign the correct name to the JSON file

for i in URI:
    
    # Print some text to make clear when iteration starts and ends:
    print("------Iteration Starts--------")
    print(i)

    # Build the URL for this iteration of the loop, print it to check what we are getting:
    URL = url_base.format(i)
    print(URL)

    # Request the html from the URL, and format as JSON:
    data = requests.get(URL).json()

    # Build the filename. Print it to check what we are getting:
    fileName = file_base.format(ons_series[j])
    print('Series we are downloading is:', ons_series[j])
    print('Data saved to', fileName)

    # Save the file:
    with open(fileName, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    
    j = j+1
```
