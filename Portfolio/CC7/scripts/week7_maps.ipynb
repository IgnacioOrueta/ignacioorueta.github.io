```python
import pandas as pd
import numpy as np
import altair as alt
from json import loads, dumps
```

### Importing data

```python
eci = pd.read_stata("../data/rankings.dta") # ECI Rankings
country_code = pd.read_csv('../data/location_country.csv') # Country codes
```

```python
eci.head()
```

```python
country_code.head()
```

### Cleaning 

```python
# Only data from 2000's
eci = eci[eci['year']>=2000]

# Merge with country code
eci_country_code = pd.merge(eci, country_code, on="country_id")

# Keeping key columns
#eci_country_code = eci_country_code[['year','name_short_en', 'iso3_code','hs_eci_rank']]

eci_country_code

# Renaming columns
eci_country_code.rename(columns={

    "name_short_en" : "country",
    "iso3_code": "ISO3"

}, inplace=True)

eci_country_code
```

### Change in the Complexity index of European countries, from the creation of the EU to now (1993 to 2022). 

```python
eci_country_code['hs_eci_1993'] = eci_country_code[eci_country_code['year'] == 1993].set_index('country_id')['hs_eci']
eci_country_code['hs_eci_2022'] = eci_country_code[eci_country_code['year'] == 2022].set_index('country_id')['hs_eci']
eci_country_code['hs_eci_1993_2022'] = (eci_country_code['hs_eci_2022'] - eci_country_code['hs_eci_1993']) / eci_country_code['hs_eci_1993']
```

```python
eci_country_code
```

### Exporting as CSV

```python
# Only 2022
eci_country_code_2022 = eci_country_code[eci_country_code['year'] ==2022]

eci_country_code_2022.to_csv('../datamap_data2022.csv', index=False)  
```
